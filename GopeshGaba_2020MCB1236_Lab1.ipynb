```python
def value_iteration(tran,states,actions, max_iterations,gamma):
    itr = 0
    value_func = []
    n = len(states)
    for i in range(n):
        value_func.append(0)
    err = 1
    eps = pow(10,-18)
    while itr<max_iterations and err>eps:
        itr+=1
        err = 0
        for state in states:
            val = []
            for action in actions:
                # print(tran[state][action])
                tv = 0
                for prob, ns, rew, is_ter in tran[state][action]:
                    tv+=prob*(rew+gamma*value_func[ns])
                    if is_ter == True:
                        print(state,action)
                        print(ns)
                val.append(tv)
            te = abs(value_func[state]-max(val))
            if te>err:
                err=te
            value_func[state] = max(val)
        print('Error for this iteration is')
        print(err)
        
    return (itr,value_func)
```

```python
def policy_iteration(tran,states,actions, max_iterations,gamma):
    n = len(states)
    
    pol = []
    for i in range(n):
        pol.append(0)
    
    stable = False
    itr = 0
    value_func = []
    
    for i in range(n):
        value_func.append(0)

    while stable==False:
        stable = True

    
        err = 1
        eps = pow(10,-15)
        while err>eps:
            
            err = 0
            for state in states:
                val = []
                action = pol[state]
                    # print(tran[state][action])
                tv = 0
                for prob, ns, rew, is_ter in tran[state][action]:
                    tv+=prob*(rew+gamma*value_func[ns])
                    if is_ter == True:
                        print(state,action)
                        print(ns)
                val.append(tv)
                te = abs(value_func[state]-max(val))
                if te>err:
                    err=te
                value_func[state] = max(val)
            print('Error for this iteration is')
            print(err)
            # itr+=0.01
        
        npol = []
        for state in states:
            pj = []
            for action in actions:
                tv = 0
                for prob, ns, rew, is_ter in tran[state][action]:
                    tv+=prob*(rew+gamma*value_func[ns])
                pj.append(tv)
            for action in actions:
                if max(pj)==pj[action]:
                    npol.append(action)
                    break
            if npol[state]!=pol[state]:
                stable=False
        pol = npol
        itr+=1

        
    return (itr,value_func)
```

```python
def dec_action(pol,state):
    p = random.random()
    i=0
    while p>pol[state][0]:
        p-=pol[state][0]
        i+=1
    return i

def MC_prediction(env,pol, episodes=1000, epsilon = 0.1,gamma=0.8):
    n = env.observation_space.n
    na = env.action_space.n
    
    Q = []
    Qf = []
    
    for i in range(n):
        
        Qs = []
        Qfa = []
        for j in range(na):
            Qsa = []
            Qfa.append(0)
            Qs.append(Qsa)
            
        Qf.append(Qfa)
        Q.append(Qs)
        
    
    ep = 0

    while ep<episodes:
        tQ = []
        for i in range(n):
            tQs = []
            for j in range(na):
                tQsa = []
                tQs.append(tQsa)
            tQ.append(tQs)
        
        observation, info = env.reset()
        events = []
        t=0
        while True:
            t+=1
            events.append(observation)
            action = dec_action(pol,observation)
            events.append(action)

            observation, reward, terminated, truncated, info = env.step(action)
            events.append(reward)

            if terminated or truncated:
                break
        G = 0
        # print(t)
        for i in range(t,0,-1):
            G = events[3*i-1]+gamma*G
            tS = events[3*i-3]
            tA = events[3*i-2]
            if len(tQ[tS][tA])==1:
                tQ[tS][tA][0]=G
            else:
                tQ[tS][tA].append(G)
        # print(G)
        for i in range(n):
            for j in range(na):
                if len(tQ[i][j])==1:
                    Q[i][j].append(tQ[i][j][0])
                if(len(Q[i][j])>0):
                    Qf[i][j]=sum(Q[i][j])/len(Q[i][j])
        

        ep+=1
    # print(Qf)
    return Qf
```

```python
env = gym.make('FrozenLake-v1', desc=None, map_name="8x8", is_slippery=True)
print(MC_prediction(env,10000,0.2))
```

```python
def cost(u):
    if u==0:
        return 0
    return 5+2*u

def poisson(k,lamda=5):
    return (math.exp(-lamda)*pow(lamda,k))/math.factorial(k)

def preward(i,j,k):
    return 8*k-cost(j)-i

n_states = 16

pt = []

for i in range(16):
    acs = []
    for j in range(16-i):
        ac = []
        qp = 0
        for k in range(i+j+1):
            if k!=i+j:
                tp = poisson(k)
                qp+=tp
                ac.append([tp,i+j-k,preward(i,j,k),False])
            else:
                tp = 1-qp
                ac.append([tp,i+j-k,preward(i,j,k),False])
        acs.append(ac)
    pt.append(acs)

# print(pt)

def value_iteration_warehouse(tran,n , max_iterations,gamma):
    itr = 0
    value_func = []
    for i in range(n):
        value_func.append(0)
    err = 1
    eps = pow(10,-18)
    while itr<max_iterations and err>eps:
        itr+=1
        err = 0
        i=0
        for state in tran:
            val = []
            for action in state:
                # print(tran[state][action])
                tv = 0
                for prob, ns, rew, is_ter in action:
                    tv+=prob*(rew+gamma*value_func[ns])
                    # if is_ter == True:
                    #     print(state,action)
                    #     print(ns)
                val.append(tv)
            te = abs(value_func[i]-max(val))
            if te>err:
                err=te
            value_func[i] = max(val)
            i+=1
        print('Error for this iteration is')
        print(err)

    npol = []
    print(len(tran))
    for state in tran:
        pj = []
        for action in state:
            tv = 0
            for prob, ns, rew, is_ter in action:
                tv+=prob*(rew+gamma*value_func[ns])
            pj.append(tv)
        jj=0
        for action in state:
            if max(pj)==pj[jj]:
                npol.append(jj)
                break
            jj+=1
    print(npol)
        
    return (itr,value_func)

print(value_iteration_warehouse(pt,n_states,100000,0.8))
```
